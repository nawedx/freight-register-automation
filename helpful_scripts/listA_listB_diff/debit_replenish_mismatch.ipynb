```python
import pandas as pd
import numpy as np
from pandas import ExcelWriter
pd.set_option('display.max_columns', 15)
df_result = pd.DataFrame(columns=['FNOT_NUMB', 'FNOT_DATE'])
```

```python
file = 'CBSP';
sheet = 'DPCB'
df = pd.read_excel(file + '.xlsx', sheet, skiprows=1)
df.columns = df.columns.str.replace(' ', '_')
df = df.dropna(subset=['FNOT_NUMB', 'FNOT_DATE'])

df['FNOT_NUMB'] = df['FNOT_NUMB'].apply(np.int64)
print(df)
df2 = df.groupby(['FNOT_NUMB', 'FNOT_DATE']).size().reset_index(name='Freq')
#print(df2)
for index, row in df2.iterrows():
    df_each = df[df.FNOT_NUMB == row['FNOT_NUMB']]
    df_each = df_each[df_each.FNOT_DATE == row['FNOT_DATE']]
    debit = df_each[df_each['CHRG_TYPE.1'] == "DEBIT"].AMOUNT.sum()
    replenish = df_each[df_each['CHRG_TYPE.1'] == "REPLENISH"].AMOUNT.sum()
    #debit = df_debit.AMOUNT.sum()
    #print(df_debit)
    if(debit != replenish):
        #print(df_each.FNOT_NUMB.iloc[0], df_each.FNOT_DATE.iloc[0].strftime('%m/%d/%Y'))
        print(df_each.FNOT_NUMB.iloc[0], df_each.FNOT_DATE.iloc[0])
        df_result = df_result.append({'FNOT_NUMB':df_each.FNOT_NUMB.iloc[0], 'FNOT_DATE':df_each.FNOT_DATE.iloc[0].strftime('%m/%d/%Y')}, ignore_index=True)
```

```python
print(df_result)
writer = ExcelWriter(sheet +'_results.xls')
df_result.to_excel(writer, 'Sheet1')
writer.save()
```
